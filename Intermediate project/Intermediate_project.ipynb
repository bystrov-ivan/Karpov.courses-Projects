#### 5.Выполните когортный анализ пользователей. 
#### В период с января по декабрь выявите когорту с самым высоким retention на 3-й месяц. Описание подхода можно найти тут. Для визуализации когортной таблицы рекомендуем использовать пример из 8-го урока python, раздел “Стильный урок”, степ 5. (15 баллов)

```python
# джойним таблицы клиентов и заказов, берем только покупки
df = orders \
            .merge(customers, on = 'customer_id') \
            .query('order_status == "delivered"') 
df.head()
```

```python
# для каждой покупки подтягиваем дату 1й покупки этим клиентом
df_min = df\
            .groupby('customer_unique_id',as_index=False)\
            .agg({'order_purchase_timestamp':'min'})

df = df.merge(df_min, on='customer_unique_id')
df.head()
```

```python
# вычленяем месяц и год из даты для формирования когорт по месяцу 1й покупки
df['cohort'] = df.order_purchase_timestamp_y.dt.strftime('%Y-%m')

# высчитываем разницу в месяцах между датой данной покупки и 1й покупки, формируя когортный период
df['m_diff'] = (df.order_purchase_timestamp_x.dt.year - df.order_purchase_timestamp_y.dt.year) * 12 \
                + (df.order_purchase_timestamp_x.dt.month - df.order_purchase_timestamp_y.dt.month)
df.head()
```

Т.к. анализ требуется с января по декабрь, такие данные имеются только за 2017 год, оставляем его. Берем данные, необходимые для анализа и аггрегируем клиентов по когортам.

```python
df_4_cohort = df\
                .query("cohort >= '2017-01' & cohort < '2018-01'") \
                .groupby(['cohort','m_diff'], as_index=False) \
                .agg({"customer_unique_id": "nunique"}) # nunique- т.к. клиент мог совершить несколько заказов за мес.

df_4_cohort.head()
```

```python
# создаем сводную
cohort = df_4_cohort.pivot(index='cohort', columns = 'm_diff', values = 'customer_unique_id')
cohort
```

```python
# считаем долю повторных клиентов,
cohort_retention = (cohort.apply(lambda x: x / cohort[0]))
cohort_retention
```

```python
# heatmap of Retenton
plt.figure(figsize=(18,10))
ax = sns.heatmap(cohort_retention,\
                 annot=True,\
                 annot_kws={"fontsize":10},\
                 vmin=0.0,\
                 vmax=0.0080,\
                 fmt='.2%',\
                 linewidths=0.01,\
                 cmap="YlGnBu")
plt.title('Retention')
plt.xlabel('Когортный период')
plt.ylabel('Когорты')
plt.yticks(rotation=45)
plt.show()
```

__Ответ: максимальный Retention на 3й месяц был 06.2017 и составлял 0.43%__

Также можно заметить, что наиболее часто клиенты возвращаются на следующий месяц,  но также влияет сезонность, и часть клиентов возвращается в осенний период и в январе

Для Frequency и  Monetary нужны кол-во заказов и сумма по ним. Т.к  каждая позиция в заказе идет отдельно, группируем по заказам, чтобы посчитать сумму каждого заказа, сразу отбираем только покупки за исследуемый год.

```python
df_4_RFM2 = df_4_RFM1\
                .query("order_purchase_timestamp >= '2017-10-18' & order_status == 'delivered'")\
                .groupby(['customer_unique_id','order_id','order_purchase_timestamp'], as_index=False)\
                .agg({'price':'sum'})
df_4_RFM2.head()
```

Считаем данные для Recency

```python
# добавляем столбец с количеством дней между покупкой и NOW
df_4_RFM2['DaysSinceOrder'] = df_4_RFM2['order_purchase_timestamp'].apply(lambda x: (NOW - x).days)
df_4_RFM2.head()
```

```python
# собимаем метрики RFM 
RFM = df_4_RFM2\
            .groupby('customer_unique_id', as_index=False)\
            .agg({'DaysSinceOrder':'min',\
                  'order_id':'count',\
                  'price':'sum'})\
            .rename(columns={'DaysSinceOrder':'Recency',\
                             'order_id':'Frequency',\
                             'price':'Monetary'})
RFM.head()
```

Для Recency и Monetary приминимо сегментирование с помощью квантилей. Для Frequency придется задать разделители вручную.
Выбрал сегментацию на 3 группы, т.к. не стоит конкретной задачи анализа, но этого должно быть достаточно для получения общей картины. Плюс для Frequency сегментация на большее кол-во не имеет большого смысла. 

```python
# Для Recency и Monetarу разделим данные на 3 группы с помощью квантилей.
quintiles = RFM[['Recency', 'Monetary']].quantile([.33, .66]).to_dict()
quintiles
```

Посмотрим детальнее на Frequency. Видно, что подавляющая часть совершает покупку только один раз, иногда клиенты возвращаются, и совсем редко приходят 3й раз и более:

```python
RFM.Frequency.value_counts()
```

```python
quintiles['Frequency'] = {.33: 1.0, .66: 2.0} 
quintiles
```

```python
#  создаем функции для  присвоения ранга 

#для Recency т.к. чем больший срок прошел, тем хуже.
def r_score(x):
    if x <= quintiles['Recency'][.33]:
        return 1
    elif x <= quintiles['Recency'][.66]:
        return 2
    else:
        return 3
    
#для Monetary и Frequency т.к. чем больше сумма или частота покупок, тем лучше.
def fm_score(x, c):
    if x <= quintiles[c][.33]:
        return 3
    elif x <= quintiles[c][.66]:
        return 2
    else:
        return 1    
```

```python
# дополняем таблицу рангами и выводим итоговую оценку путем объединения всех рангов
RFM['R'] = RFM['Recency'].apply(lambda x: r_score(x))
RFM['F'] = RFM['Frequency'].apply(lambda x: fm_score(x, 'Frequency'))
RFM['M'] = RFM['Monetary'].apply(lambda x: fm_score(x, 'Monetary'))
RFM['RFM Score'] = RFM['R'].map(str) + RFM['F'].map(str) + RFM['M'].map(str)
RFM.head()
```

Сегментация:
- Ядро — или стратегически важные клиенты. Покупают часто и на большие суммы. Хотят чувствовать исключительность своего положения: закрытый клуб, золотая карта — ценят вещи, которые показывают особый статус (111).

- Лояльные клиенты: тратят немного, но часто. Нуждаются в постоянной поддержке контакта, бонусах, подарках при заказе (112, 122, 113, 123).

- «Киты»: тратят не слишком часто, но много. Скидки их интересуют в меньшей степени, чем ценный, возможно эксклюзивный товар (121).

- Новички: могут, как уйти в категорию спящих, так и в категорию лояльных. Требуют пристального внимания, помощи в выборе товара, полезном контенте (132, 133).

- «Сони»: когда-то были хорошими покупателями, но почему-то перестали проявлять активность. Можно попробовать вернуть былую свежесть вашим отношениям: предложить новый продукт, рассказать об акциях, выяснить, почему они потеряли интерес (211, 212, 213, 221, 222, 223, 231, 232, 233).

- Уходящие: одной ногой уже за порогом. Следует узнать, почему не хотят пользоваться вашими услугами, предложить выгодную акцию, начислить бонусы, подарить промокод (311, 312, 312, 321, 322, 323, 331, 332).

- Потерянные клиенты. Смириться и отпустить (333). 

```python
segt_map = {
    r'111': 'Ядро',
    r'1[1-2][2-3]': 'Лояльные клиенты',
    r'121': 'Киты',
    r'13[1-3]': 'Новички',
    r'2[1-3][1-3]': 'Сони',
    r'3[1-3][1-3]' : 'Уходящие',
    r'333': 'Потерянные клиенты'
}

RFM['Segment'] = RFM['RFM Score'].map(str)
RFM['Segment'] = RFM['Segment'].replace(segt_map, regex=True)
RFM.head()
```
